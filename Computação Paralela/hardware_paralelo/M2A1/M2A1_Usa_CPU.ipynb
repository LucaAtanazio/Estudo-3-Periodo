```python
import multiprocessing
import os
import math
import time

def noise_maker(id):
    # Ocupa o núcleo com cálculos pesados que não usam muita memória,
    # apenas forçam a unidade de execução da CPU.
    print(f"Vilão {id} iniciando no PID: {os.getpid()}")
    while True:
        x = math.sqrt(math.exp(math.log(1234.5678)))
        x = x * x

if __name__ == "__main__":
    # Criamos 6 processos para "roubar" tempo de CPU
    viloes = []
    for i in range(6):
        p = multiprocessing.Process(target=noise_maker, args=(i,))
        p.start()
        viloes.append(p)
    
    print("\n--- 6 Núcleos estão sendo ocupados agora! ---")
    print("Abra o Gerenciador de Tarefas (Ctrl+Shift+Esc) e veja o uso da CPU.")
    print("Para parar, interrompa o Kernel do Notebook.")
```
